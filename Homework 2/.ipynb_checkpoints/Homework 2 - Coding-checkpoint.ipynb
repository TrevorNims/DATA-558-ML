## Derivations

In this first section, I will perform a couple of important derivations on our objective $F(\beta)$:

$$F(\beta) = \frac{1}{2n}\sum_{i=1}^{n}(y_i-x_i^T\beta)^2+\frac{\lambda}{n}||\beta||_2^2$$

First, under the assumption of $d = 1$ and $n = 1$, I will find the derivatie of our objective $F(\beta)$.

$$
\begin{align*}
\frac{\partial}{\partial \beta}F(\beta) &= \frac{\partial}{\partial \beta}(\frac{1}{2}(y-x\beta)^2+\lambda \beta^2) \\
&= -x(y-x\beta) + 2\lambda \beta
\end{align*}
$$

Now, under the assumption of $d > 1$ and $n > 1$, I will find the derivatie of our objective $F(\beta)$.

$$
\frac{\partial}{\partial \beta}F(\beta) = \frac{\partial}{\partial \beta}(\frac{1}{2n}||Y-X^T\beta||_2^2)+\frac{\partial}{\partial \beta}(\frac{\lambda}{n}||\beta||_2^2)
$$
Let us name the first addend $f(\beta)$ and the second $g(\beta)$, then we have:
$$
\begin{align*}
\frac{\partial}{\partial \beta}f(\beta) &= \frac{\partial}{\partial \beta}(\frac{1}{2n}(Y-X^T\beta)^T(Y-X^T\beta)) \\
&= \frac{\partial}{\partial \beta} (\frac{1}{2n}(Y^TY-Y^TX^T\beta-\beta^TXY+\beta^TXX^T\beta)) \\
&= \frac{1}{2n}(-\frac{\partial}{\partial \beta}(Y^TX^T\beta)-\frac{\partial}{\partial \beta}(\beta^TXY)+\frac{\partial}{\partial \beta}(\beta^TXX^T\beta)) \\
&= \frac{1}{2n}(-XY-XY + (XX^T + (XX^T)^T)\beta) \\
&= -\frac{X}{n}(Y-X^T\beta)
\end{align*}
$$
Now consider $g(\beta)$:
$$
\begin{align*}
\frac{\partial}{\partial \beta}g(\beta) &= \frac{\partial}{\partial \beta}(\frac{\lambda}{n}\beta^T\beta) \\
&= (\frac{2\lambda}{n})\beta
\end{align*}
$$
Therefore, the gradient is given by the sum $f(\beta)+g(\beta)$, that is:
$$\frac{\partial}{\partial \beta}F(\beta) = -\frac{X}{n}(Y-X^T\beta)+(\frac{2\lambda}{n})\beta$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
# Load the data

file = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(file, sep=',', header=0)
penguins = penguins.dropna()
```

```python
# Create our X matrix with the predictors and y vector with the response

X = penguins.drop('body_mass_g', axis=1)
X = pd.get_dummies(X, drop_first=True)
y = penguins['body_mass_g']
```

```python
# Divide the data into training and test sets. By default, 25% goes into the test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Standardize all values

scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

scaler = preprocessing.StandardScaler().fit(y_train.values.reshape(-1,1))
y_train = scaler.transform(y_train.values.reshape(-1,1)).reshape((-1))
y_test = scaler.transform(y_test.values.reshape(-1,1)).reshape((-1))
```

```python
# Define gradient function

def compute_grad(beta, X, y, lam):
    n = len(y)
    x_tb = np.dot(np.transpose(X), beta)
    return -1/n*np.dot(X, y-x_tb)+2*lam/n*beta
```

```python
# Define gradient descent function

def grad_descent(beta_init, step_size, num_iter, *args):
    iterates = []
    beta = beta_init
    for i in range(num_iter):
        new_beta = beta - step_size*compute_grad(beta, args[0], args[1], args[2])
        iterates.append(new_beta)
        beta = new_beta
    return iterates
```

```python
# Run 1000 iterations of gradient decent on training data with lambda = -5.0

beta_init = np.zeros(X_train.shape[1])
step_size=0.5
iterations = 1000
lam = -5.0

iterates = grad_descent(beta_init, step_size, iterations, np.transpose(np.array(X_train)), np.array(y_train), lam)
```

```python
# Define objective function

def obj_fun(beta, X, y, lam):
    n = len(y)
    x_tb = np.dot(np.transpose(X), beta)
    return 1/2*n*np.dot(np.transpose(y-x_tb), y-x_tb)+lam/n*np.dot(np.transpose(beta), beta)
```

```python
# Plot value of objective function vs. iteration counter

points = []
for i in range(iterations):
    points.append(obj_fun(iterates[i], np.transpose(np.array(X_train)), np.array(y_train), lam))
plt.xlabel('Step Number ($t$)')
plt.ylabel(r'Value of Objective Function $F(\beta_t)$')
plt.title(r'Value of $F(\beta_t)$ vs $t$ ($\lambda=-5.0$)')
plt.plot(points)
```

As can be seen above, the local minimum of the objective function is achieved very quickly with a negative lambda value. As the algorithm continues past this minimum, the objective function appears to increase monotonically.

```python
# Run 1000 iterations of gradient decent on training data with lambda = .05, plot values of
# objective function vs. iteration counter

step_size=0.5
iterations = 1000
lam = .05

iterates = grad_descent(beta_init, step_size, iterations, np.transpose(np.array(X_train)), np.array(y_train), lam)

points = []
for i in range(iterations):
    points.append(obj_fun(iterates[i], np.transpose(np.array(X_train)), np.array(y_train), lam))
plt.xlabel('Step Number ($t$)')
plt.ylabel(r'Value of Objective Function $F(\beta_t)$')
plt.title(r'Value of $F(\beta_t)$ vs $t$ ($\lambda=.05$)')
plt.plot(points)
```

As can be seen above, the local minimum of the objective function is achieved very quickly with a postive lambda value as well. In contrast to the algorithm's performance with a negative lambda value, however, this choice of lambda does not cause the objective function to monotonically increase after reaching the local minimum. Instead, it actually appears to decrease monotonically.

```python
# Define sklearn's objective function

def obj_fun_sk(beta, X, y, alpha):
    n = len(y)
    x_tb = np.dot(np.transpose(X), beta)
    return np.dot(np.transpose(y-x_tb), y-x_tb)+alpha*np.dot(np.transpose(beta), beta)
```

```python
# Train and compute coefficients of sklearn model with alpha = lambda/n, compare coefficients from my model

alph = lam/len(y_train)

mod = Ridge(alpha=alph).fit(X_train,y_train)
print('Coefficients from Trevor\'s implementation:\n', iterates[-1])
print('\nCoefficients from sklearn\'s implementation:\n', [round(float(i), 8) for i in mod.coef_])
```

As can be seen above, the coefficients are very similar. Despite scaling the penalization term, it's interesting that without scaling the empirical risk term I still achieved very similar results. This almost leads me to believe that scaling that term may not have a huge effect on the coefficient optimization.

```python
# Compare objective function's value

print('Objective Function Value for Trevor\'s implementation:\n',
      obj_fun(iterates[-1], np.transpose(np.array(X_train)), np.array(y_train), lam))
print('\nObjective Function Value for sklearn\'s implementation:\n',
      obj_fun_sk(mod.coef_, np.transpose(np.array(X_train)), np.array(y_train), alph))
```

That being said, the objective function values differ significantly.

```python
# Run gradient descent over logspace of step-size values with lambda=.05, save the final iterate as well as
# the objective function's value when given the final iterate for each iteration.

iterations = 1000
lam = .05

eta_vals = np.logspace(-5,-.4,20)
coef_list = []
obj_vals = []

for eta in eta_vals:
    step_size = eta
    coefs = grad_descent(beta_init, step_size, iterations, np.transpose(np.array(X_train)), np.array(y_train), lam)
    coef_list.append(coefs[-1])
    obj_vals.append(obj_fun(coefs[-1], np.transpose(np.array(X_train)), np.array(y_train), lam))
```

```python
# Plot results (for fun!)

plt.plot(np.log(eta_vals), obj_vals)
plt.xlabel(r'$log(\eta)$')
plt.ylabel(r'Value of Objective Function $F(\beta_t)$')
plt.title(r'Value of $F(\beta_t)$ vs $log(\eta)$ ($\lambda=.05$)')
plt.show()
```

```python
# Print optimal eta value and coefficients

print('OPTIMAL STEP SIZE eta = ', eta_vals[np.argmin(obj_vals)])

print('\nCoefficients from Trevor\'s improved implementation:\n', coef_list[np.argmin(obj_vals)])
print('\nCoefficients from sklearn\'s implementation:\n', [float(i) for i in mod.coef_])
```

Again, the coefficients appear very similar. It appears that by decreasing the original step size slightly we arrive at a more optimal solution.

```python
# Compare my objective function's value with that of sklearn

print('Objective Function Value for Trevor\'s implementation:\n',
      obj_fun(coef_list[np.argmin(obj_vals)], np.transpose(np.array(X_train)), np.array(y_train), lam))
print('\nObjective Function Value for sklearn\'s implementation:\n',
      obj_fun_sk(mod.coef_, np.transpose(np.array(X_train)), np.array(y_train), alph))
```

As expected, the objective function values vary greatly.

```python
# Define gradient descent function with additional epsilon parameter

def grad_descent_improved(init_p, step_size, epsilon, *args):
    iterates = []
    beta = beta_init
    grad = compute_grad(beta, args[0], args[1], args[2])
    while np.linalg.norm(grad) > epsilon:
        new_beta = beta - step_size*compute_grad(beta, args[0], args[1], args[2])
        iterates.append(new_beta)
        beta = new_beta
        grad = compute_grad(beta, args[0], args[1], args[2])
    return iterates
```

```python
# Run gradient decent with epsilon = .005, lambda = .05 on training data, plot values of
# objective function vs. iteration counter

step_size=0.5
epsilon = .005
lam = .05

iterates = grad_descent_improved(beta_init, step_size, epsilon, 
                                 np.transpose(np.array(X_train)), np.array(y_train), lam)

points = []
for i in range(len(iterates)):
    points.append(obj_fun(iterates[i], np.transpose(np.array(X_train)), np.array(y_train), lam))
plt.xlabel('Step Number ($t$)')
plt.ylabel(r'Value of Objective Function $F(\beta_t)$')
plt.title(r'Value of $F(\beta_t)$ vs $t$ ($\lambda=.05$)')
plt.plot(points)
```

```python
# Compare coefficients from my revised model with those of sklearn's

print('Coefficients from Trevor\'s "improved" grad-desc implementation:\n', iterates[-1])
print('\nCoefficients from sklearn\'s implementation:\n', [round(float(i), 8) for i in mod.coef_])
```

Despite finishing in less iterations, with $\varepsilon$ fixed at 0.005 it appears my "improved" gradient descent algorithm may actually perform slightly worse than my original approach.

```python
# Compare objective function's value

print('Objective Function Value for Trevor\'s "improved" grad-desc implementation:\n',
      obj_fun(iterates[-1], np.transpose(np.array(X_train)), np.array(y_train), lam))
print('\nObjective Function Value for sklearn\'s implementation:\n',
      obj_fun_sk(mod.coef_, np.transpose(np.array(X_train)), np.array(y_train), alph))
```

With an objective function value higher than that of the original implementation, this further corroborates the notion that $\varepsilon$ may need to be decreased to acheive a more optimal result.

```python
# Run gradient descent over logspace of step-size values with lambda=.05, epsilon=.005, save
# the final iterate as well as the objective function's value when given the final iterate for each iteration.

epsilon = .005
lam = .05

eta_vals = np.logspace(-5,-.4,20)
coef_list = []
obj_vals = []

for eta in eta_vals:
    step_size = eta
    coefs = grad_descent_improved(beta_init, step_size, epsilon, 
                                  np.transpose(np.array(X_train)), np.array(y_train), lam)
    coef_list.append(coefs[-1])
    obj_vals.append(obj_fun(coefs[-1], np.transpose(np.array(X_train)), np.array(y_train), lam))
```

As can be seen above, I used a keyboard interrupt to stop the loop from continuing. It appeared that with very small step sizes (take the first value in 'eta_vals' of 1e-6 for example), the norm of the gradient may take an incredibly large - if not infinite - amount of time to shrink below a given stopping criterion $\varepsilon$.

```python
# Run gradient descent over logspace of MUCH LARGER step-size values with lambda=.05, epsilon=.005, save
# the final iterate as well as the objective function's value when given the final iterate for each iteration.

epsilon = .005
lam = .05

eta_vals = np.logspace(-1,-.4,20)
coef_list = []
obj_vals = []

for eta in eta_vals:
    step_size = eta
    coefs = grad_descent_improved(beta_init, step_size, epsilon, 
                                  np.transpose(np.array(X_train)), np.array(y_train), lam)
    coef_list.append(coefs[-1])
    obj_vals.append(obj_fun(coefs[-1], np.transpose(np.array(X_train)), np.array(y_train), lam))
```

```python
# Plot results (for fun!)

plt.plot(np.log(eta_vals), obj_vals)
plt.xlabel(r'$log(\eta)$')
plt.ylabel(r'Value of Objective Function $F(\beta_t)$')
plt.title(r'Value of $F(\beta_t)$ vs $log(\eta)$ ($\lambda=.05$)')
plt.show()
```

```python
# Print optimal eta value and coefficients, compare my objective function's value with that of sklearn

print('OPTIMAL STEP SIZE eta = ', eta_vals[np.argmin(obj_vals)])

print('\nCoefficients from Trevor\'s "improved" implementation:\n', coef_list[np.argmin(obj_vals)])
print('\nCoefficients from sklearn\'s implementation:\n', [round(float(i), 8) for i in mod.coef_])

print('\nObjective Function Value for Trevor\'s "improved" implementation:\n',
      obj_fun(coef_list[np.argmin(obj_vals)], np.transpose(np.array(X_train)), np.array(y_train), lam))
print('\nObjective Function Value for sklearn\'s implementation:\n',
      obj_fun_sk(mod.coef_, np.transpose(np.array(X_train)), np.array(y_train), alph))
```

The interesting takeaway here is, in my opinion, better illustrated in the plot above. Despite arriving at the same optimal value of lambda, as can be seen, our revised algorithm is not monotonically decreasing (as we saw in our original implementation), and instead looks highly variable.

```python
import statsmodels.api as sm
```

```python
file = 'https://www.statlearning.com/s/Auto.csv'
auto = pd.read_csv(file, sep=',', na_values='?')
auto = auto.dropna()
auto.head()
```

```python
# Define X and y, create and fit model, print summary

X = np.array(auto['weight'])
y = np.array(auto['mpg'])

X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

**i)** With a non-zero coefficient with statistical significance, it appears there does exist a relationship between vehicle weight and mpg.

**ii)** With an R-squared value of .693, I would say that the relationship is pretty strong. 69.3% of the variability in mpg can be explained by vehicle weight.

**iii)** As can be seen above, the coefficient for weight is negative. Therefore, in general, the heavier the car the lower the mpg.

```python
fig, ax = plt.subplots()
sm.graphics.plot_fit(results, 1, ax=ax)
ax.set_ylabel("mpg")
ax.set_xlabel("weight")
ax.set_title("Car Weight vs MPG")
```

```python
fig = plt.figure(figsize=(10, 8))
sm.graphics.plot_regress_exog(results, 1, fig=fig)
fig.tight_layout(pad=1.0)
```

As can be seen in the top-right pot above, the residuals don't seem to be randomly distributed across the fitted values axis. This is concerning as it suggests that our model may have been too inflexible (biased) to capture some of the intrinsic complexity of the data.
